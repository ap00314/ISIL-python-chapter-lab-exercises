**2.3.5 Additional Graphical and Numerical Summaries**

以下代码对应原书中的：
> plot(Auto$cylinders , Auto$mpg)

> attach (Auto)

> plot(cylinders , mpg)

```python
plt.figure(figsize = (10,6))
plt.scatter(Auto.cylinders,Auto.mpg,color = '',edgecolor = 'black',linewidths = 0.7)
plt.xticks(fontsize = 12)
plt.yticks(np.linspace(10,40,4),fontsize = 12)
plt.xlabel('cylinders',fontsize = 18,labelpad = 16)
plt.ylabel('mpg',fontsize = 18,labelpad = 16)
plt.show()
```

```python
#对应原书中plot(cylinders , mpg, col="red", varwidth =T, xlab="cylinders ",ylab="MPG")
plt.figure(figsize = (10,6))
cylinders = np.sort(Auto.cylinders.unique())
data_cylinders = [Auto.mpg[Auto.cylinders == cylinder] for cylinder in cylinders]
data_widths = [(Auto.cylinders == cylinder).sum()/len(Auto) + 0.2 for cylinder in cylinders]
bp = plt.boxplot(data_cylinders,sym = 'o', patch_artist = True ,widths=data_widths,labels=cylinders)
for box in bp['boxes']:
    box.set(facecolor = 'red', linewidth=1)        # 箱体边框颜色
for whisker in bp['whiskers']:
    whisker.set(color='k', linewidth=0.5,linestyle='-.')
for cap in bp['caps']:
    cap.set(color='Black', linewidth=1)
for median in bp['medians']:
    median.set(color='Black', linewidth=2)
for flier in bp['fliers']:
    flier.set(marker='o',color='Black',linewidth=0.5,alpha = 1)
plt.xticks(fontsize = 12)
plt.yticks(np.linspace(10,40,4),fontsize = 12)
plt.xlabel('cylinders',fontsize = 18,labelpad = 16)
plt.ylabel('mpg',fontsize = 18,labelpad = 16)
plt.show()
```

```python
# 对应原书中hist(mpg,col=2,breaks =15)
f = Auto.hist('mpg',bins = 20,range = (5,60),color='Red',edgecolor = 'black',figsize = (10,6))
plt.grid(linestyle='')
plt.title('')
plt.xticks(np.linspace(10,40,4),fontsize = 12)
plt.yticks(np.linspace(0,40,5),fontsize = 12)
plt.xlabel('mpg',fontsize = 18,labelpad = 16)
plt.ylabel('Frequency',fontsize = 18,labelpad = 16)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 10))
plt.show()
```
