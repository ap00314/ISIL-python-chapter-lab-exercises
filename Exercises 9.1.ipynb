```python
# To support both python 2 and python 3
# 让这份笔记同步支持 python 2 和 python 3
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

**1. This problem involves hyperplanes in two dimensions.
(a) Sketch the hyperplane 1 + 3X 1 − X 2 = 0. Indicate the set of
points for which 1 + 3X 1 − X 2 > 0, as well as the set of points
for which 1 + 3X 1 − X 2 < 0.
(b) On the same plot, sketch the hyperplane −2 + X 1 + 2X 2 = 0.
Indicate the set of points for which −2+ X 1 + 2X 2 > 0, as well
as the set of points for which −2 + X 1 + 2X 2 < 0.**

```python
plt.figure(figsize=(7, 6))
XX, YY = np.meshgrid(np.arange(-5,10,5), np.arange(-20,30,10))
X1_min,X1_max = -11,11
plt.plot([X1_min,X1_max],[X1_min * 3 + 1,X1_max * 3 + 1],'k-',label = '1 + 3$X_1$ - $X_2$ = 0')
plt.plot([X1_min,X1_max],[1 - X1_min/2,1 - X1_max/2],'g-',label = '-2 + $X_1$ + 2$X_2$ = 0')
z = (3 * XX.ravel() - YY.ravel() + 1> 0)
plt.scatter(XX.ravel()[z],YY.ravel()[z],marker = 'o',s = 200 ,edgecolors  = 'red',color = '',label = '1 + 3$X_1$ - $X_2$ > 0')
z = (3 * XX.ravel() - YY.ravel() + 1< 0)
plt.scatter(XX.ravel()[z],YY.ravel()[z],marker = 'o',s = 200 ,edgecolors  = 'blue',color = '',label = '1 + 3$X_1$ - $X_2$ > 0')
z = (XX.ravel() + 2 * YY.ravel() - 2> 0)
plt.scatter(XX.ravel()[z],YY.ravel()[z],marker = '|',s = 200 ,color ='red',label = '-2 + $X_1$ + 2$X_2$ > 0')
z = (XX.ravel() + 2 * YY.ravel() - 2< 0)
plt.scatter(XX.ravel()[z],YY.ravel()[z],marker = '_',s = 200 ,color ='blue',label = '-2 + $X_1$ + 2$X_2$ < 0')
plt.xlim([-10,10])
plt.ylim([-30,30])
plt.xlabel('$X_1$',fontsize = 16,labelpad =10)
plt.ylabel('$X_2$',fontsize = 16,labelpad =10)
plt.legend()
plt.show()
```
