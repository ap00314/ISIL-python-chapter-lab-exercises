6. We will now explore (6.12) and (6.13) further.
(a) Consider (6.12) with p = 1. For some choice of y 1 and λ > 0,
plot (6.12) as a function of β 1 . Your plot should confirm that
(6.12) is solved by (6.14).

```python
y = 2
lambda_ = 2
betas = np.arange(-10, 10.1, 0.1)
func = (y - betas) ** 2 + lambda_ * betas ** 2
plt.plot(betas, func,':')
beta_ckecking = y/(1 + lambda_)
func_checking = (y - beta_ckecking) ** 2 + lambda_ * beta_ckecking ** 2
plt.plot(beta_ckecking, func_checking, marker = 'o',color = 'r')
plt.xlabel(r"$\beta$", fontsize=18)
plt.ylabel(u"Ridge optimization", fontsize=18)
plt.show()
```

(b) Consider (6.13) with p = 1. For some choice of y 1 and λ > 0,
plot (6.13) as a function of β 1 . Your plot should confirm that
(6.13) is solved by (6.15).

```python
# y > lambda / 2
func = (y - betas) ** 2 + lambda_ * abs(betas)
plt.plot(betas, func,':')
beta_ckecking = y - lambda_/2
func_checking = (y - beta_ckecking) ** 2 + lambda_ * abs(beta_ckecking)
plt.plot(beta_ckecking, func_checking, marker = 'o',color = 'r')
plt.xlabel(r"$\beta$", fontsize=18)
plt.ylabel(u"Ridge optimization", fontsize=18)
plt.show()
```
