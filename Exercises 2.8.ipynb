 **iii. Use the plot() function to produce side-by-side boxplots of
Outstate versus Private**

```python
#对应原书中plot(Private , Outstate, col="red", varwidth =T, xlab="cylinders ",ylab="MPG")
plt.figure(figsize = (6,6))
Privates = np.sort(college.Private.unique())
data_Privates = [college.Outstate[college.Private == Private] for Private in Privates]
data_widths = [(college.Private == Private).sum()/len(college)  for Private in Privates]
bp = plt.boxplot(data_Privates,sym = 'o', patch_artist = True ,widths=data_widths,labels=Privates)
for box in bp['boxes']:
    box.set(facecolor = 'red', linewidth=1)        # 箱体边框颜色
for whisker in bp['whiskers']:
    whisker.set(color='k', linewidth=0.5,linestyle='-.')
for cap in bp['caps']:
    cap.set(color='Black', linewidth=1)
for median in bp['medians']:
    median.set(color='Black', linewidth=2)
for flier in bp['fliers']:
    flier.set(marker='o',color='Black',linewidth=0.5,alpha = 1)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Private',fontsize = 18,labelpad = 16)
plt.ylabel('Outstate',fontsize = 18,labelpad = 16)
plt.show()
```

**iv. Create a new qualitative variable, called Elite , by binning
the Top10perc variable. We are going to divide universities
into two groups based on whether or not the proportion
of students coming from the top 10% of their high school
classes exceeds 50%.Use the summary() function to see how many elite univer-
sities there are. Now use the plot() function to produce
side-by-side boxplots of Outstate versus Elite**

以下类似于原文代码：

> Elite=rep("No",nrow(college))

> Elite[college$Top10perc >50]=" Yes"

> Elite=as.factor(Elite)

> college=data.frame(college ,Elite)

```python
Elite = college['Top10perc'].apply(lambda x:'Yes' if x > 50 else 'No' )
college['Elite'] = Elite
```

```python
college
```

Use the summary() function to see how many elite univer- sities there are

```python
#同上的另一种办法
college['Elite'].value_counts()
```

 Now use the plot() function to produce
side-by-side boxplots of Outstate versus Elite

```python
#对应原书中plot(Elite , Outstate, col="red", varwidth =T, xlab="cylinders ",ylab="MPG")
plt.figure(figsize = (6,6))
Elites = np.sort(college.Elite.unique())
data_Elites = [college.Outstate[college.Elite == Elite] for Elite in Elites]
data_widths = [(college.Elite == Elite).sum()/len(college)  for Elite in Elites]
bp = plt.boxplot(data_Elites,sym = 'o', patch_artist = True ,widths=data_widths,labels=Elites)
for box in bp['boxes']:
    box.set(facecolor = 'red', linewidth=1)        # 箱体边框颜色
for whisker in bp['whiskers']:
    whisker.set(color='k', linewidth=0.5,linestyle='-.')
for cap in bp['caps']:
    cap.set(color='Black', linewidth=1)
for median in bp['medians']:
    median.set(color='Black', linewidth=2)
for flier in bp['fliers']:
    flier.set(marker='o',color='Black',linewidth=0.5,alpha = 1)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Elite',fontsize = 18,labelpad = 16)
plt.ylabel('Outstate',fontsize = 18,labelpad = 16)
plt.show()
```

**v. Use the hist() function to produce some histograms with

ables. You may find the command par(mfrow=c(2,2)) useful:
